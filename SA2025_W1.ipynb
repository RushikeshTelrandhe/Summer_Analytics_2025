### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()
above_median_mpg_cars = df[df['mpg'] > median_mpg].copy()
above_median_mpg_cars.reset_index(drop=True, inplace=True)
above_median_mpg_cars['hp_to_weight_ratio'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']
max_ratio = above_median_mpg_cars['hp_to_weight_ratio'].max()
best_ratio_cars = above_median_mpg_cars[above_median_mpg_cars['hp_to_weight_ratio'] == max_ratio]
print("Car(s) with the best horsepower to weight ratio among those with above-median mpg:")
for index, row in best_ratio_cars.iterrows():
    print(row['name'])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin')
plt.title('Evolution of Average MPG by Year and Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
car_year_counts = df.groupby('name')['model_year'].nunique()
multi_year_cars = car_year_counts[car_year_counts > 1].index
multi_year_df = df[df['name'].isin(multi_year_cars)].copy()
mpg_std = multi_year_df.groupby('name')['mpg'].std()
consistent_cars = mpg_std[mpg_std < 1.0].index
consistent_df = multi_year_df[multi_year_df['name'].isin(consistent_cars)].copy()
consistent_cars_summary = consistent_df.groupby('name').agg(
    appearances=('model_year', 'count'),
    average_mpg=('mpg', 'mean')
).reset_index()
lowest_variation_std = mpg_std[consistent_cars].min()
lowest_variation_cars = mpg_std[mpg_std == lowest_variation_std].index
result = consistent_cars_summary[consistent_cars_summary['name'].isin(lowest_variation_cars)]
result_sorted = result.sort_values(by=['appearances', 'average_mpg'], ascending=[False, False])
print("Car(s) with the lowest variation in mpg (< 1.0) across multiple years:")
print(result_sorted)
```
